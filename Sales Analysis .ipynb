```python
import pandas as pd
import os
```

```python
#merging 12 months of sales data into one CSV file
```

```python
df=pd.read_csv('./Sales_Data/Sales_April_2019.csv')
all_month_data=pd.DataFrame()
files=[file for file in os.listdir('./Sales_Data')]
for file in files:
    df=pd.read_csv('./Sales_Data/'+file)
    all_month_data=pd.concat([all_month_data,df])
all_month_data.head()
```

```python
all_month_data.to_csv('All_data.csv',index=False)
```

```python
all_data=pd.read_csv('All_data.csv')
all_data.head()
```

```python
all_data.shape
```

Add month columns

```python
all_data=all_data.dropna(how='all')
all_data.head()
```

```python
all_data.shape
```

```python
all_data=all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
all_data['Month']=all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
all_data['Sales']=all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
results=all_data.groupby('Month').sum()
print(results)
```

```python
import matplotlib.pyplot as plt
```

```python
month = range(1,13)
plt.bar(month,results['Sales'])
plt.xticks(month)
plt.ylabel('Sales in USD ($)')
plt.xlabel('month number')
```

```python
all_data['City']=all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data.head()
```

```python
results=all_data.groupby(['City']).sum()
print(results)
```

```python
cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('Sales in USD($)')
plt.xlabel('City name')
plt.show()
```

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data.head()
```

```python
hour=[hour for hour, df in all_data.groupby('Hour')]
plt.plot(hour,all_data.groupby(['Hour']).count())
all_data.groupby(['Hour']).count()
plt.xticks(hour)
plt.xlabel("Hour")
plt.ylabel('Number of order')
plt.grid()
plt.show()
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped']=df.groupby("Order ID")['Product'].transform(lambda x: ','.join(x))
df=df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common(10)
```

```python
product_group=all_data.groupby('Product')
quantity_order=product_group.sum()['Quantity Ordered']
products=[product for product, df in product_group]
plt.bar(products,quantity_order)
plt.ylabel('Quantity Ordered')
plt.xlabel('product')
plt.xticks(products,rotation='vertical',size=10)
plt.show
```

```python
prices=all_data.groupby('Product').mean()['Price Each']
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity_order,color='g')
ax2.plot(products,prices,'b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('price($)',color='b')
ax1.set_xticklabels(products,rotation='vertical',size=8)
plt.show()
```
